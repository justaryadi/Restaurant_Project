### 3.2 Data cleaning

```python
def general_cleaning(df):
    # Kecilin semua huruf dan hapus spasi di awal/akhir 
    df['name'] = df['name'].str.lower().str.strip()
    df['address'] = df['address'].str.lower().str.strip()
    df['online_order'] = df['online_order'].str.lower().str.strip()
    df['book_table'] = df['book_table'].str.lower().str.strip()
    df['location'] = df['location'].str.lower().str.strip()
    df['rest_type'] = df['rest_type'].str.lower().str.strip()
    df['dish_liked'] = df['dish_liked'].str.lower().str.strip()
    df['cuisines'] = df['cuisines'].str.lower().str.strip()
    df['listed_in(type)'] = df['listed_in(type)'].str.lower().str.strip()
    df['listed_in(city)'] = df['listed_in(city)'].str.lower().str.strip()

general_cleaning(df)
```

##### fungtion general cleaning
supaya tidak semrawut antara huruf kecil dan besar, serta ada spasi lebih makanya diseragamkan hurufnya jadi kecil semua dan hapus spasi lebihnya.

### 3.3 Data Spesific Cleaning

```python
#semua jadi string, lalu ambil karakter sebelum tanda '/'
df['rate'] = df['rate'].astype(str).str.split('/').str[0]
#ganti teks yang bukan angka jadi NaN (kosong) supaya bisa dihitung nntinya
#ganti 'NEW', 'new', dan tanda '-' 
df['rate'] = df['rate'].replace(['NEW', 'new', '-', 'nan'], pd.NA)
#ubahjadi angka (float)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'].astype(str).str.replace(',', '', regex=False).str.strip(),errors='coerce').astype('Int64')
```

```python
#pastikan jadi string dulu supaya bisa diolah, kalo langsung angka bisa error
df['phone'] = df['phone'].astype(str)
#ambil bagian depan sebelum ada tanda baca/pemisah
#split berdasarkan karakter non-angka yang sering jadi pemisah
df['phone'] = df['phone'].str.split(r'[,/\n]').str[0]
#bersihkan karakter (spasi, +, tanda kurung),r'\D' hapus semua yang bukan angka
df['phone'] = df['phone'].str.replace(r'\D', '', regex=True)
df['phone'] = df['phone'].replace('', np.nan)
df['phone'] = df['phone'].fillna('Unknown')  # Isi data kosong dengan 'Unknown'
```

yang di bagian phone itu sangat kotor jadi di seragamkan menjadi angka aja tanpa ada simnol lain, dan menyesuaikan di tipe datanya.

```python
# hapus duplikat berdasarkan Nama, Alamat, dan Lokasi
df.drop_duplicates(subset=['name', 'address', 'location'], keep='first', inplace=True)
```

supaya tidak ada yang muyncul berkali-kali, dan supaya di bagian rating gk ancur nantinya karna ada yang double. 

```python
# hapus baris yang location dan cuisines yang kosongnya sedikit
df.dropna(subset=['location', 'cuisines'], inplace=True)
```

menghapus /5 dan menyesuaikan tipe datanya supaya terlihat rapi dan sesuai tipe datanya.

```python
df_med = df['rate'].median()
df['rate'] = df['rate'].fillna(df_med)
```

```python
df_med = df['approx_cost(for two people)'].median()
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df_med)
```

kenapa pakai median untuk menambal yg kosong di "rate" dan "aprox_cost" supaya gk outlier lebih

```python
df_modus = df['rest_type'].mode()[0]  #ambil nilai modus pertama
print("Modus rest_type:", df_modus)

df['rest_type'] = df['rest_type'].fillna(df_modus)
```

```python
df_modus = df['dish_liked'].mode() #menampilkan modus pada kolom dish_liked

df['dish_liked'] = df['dish_liked'].fillna(df_modus[0])  #mengisi data yang kosong dengan modus pada kolom dish_liked
```

kenapa pakai modus untuk menambal yg kosong di "rest_type dan dish_liked" supaya ttahu rempat yang paling sering muncul/dipilih

### 3.4 Data Manipulation and Future Engineering

```python
import ast

#Ubah teks panjang jadi jumlh (Angka)
df['total_reviews'] = df['reviews_list'].apply(lambda x: len(ast.literal_eval(x)) if isinstance(x, str) else 0)

#Ubah daftar menu jadi status (0 atau 1)
df['has_menu'] = df['menu_item'].apply(lambda x: 0 if x == '[]' else 1)

#hapus kolom aslinya
df.drop(columns=['reviews_list', 'menu_item'], inplace=True)
```


##### Mengetahui seberapa populer restoran dari banyaknya review, Membedakan restoran yang punya menu digital vs yang tidak, hapus kolom asli meni_item supaya ringanfile

```python
# Cek rata-rata rating dan jumlah review berdasarkan lokasi
summary = df.groupby('location').agg({
    'rate': 'mean',
    'total_reviews': 'sum',
    'name': 'count'
}).rename(columns={'name': 'restaurant_count'}).sort_values(by='rate', ascending=False)

print(summary.head(10))
```

#### dengan tujuan mengatahui valuasi dari restoran di lokasi yang berbeda

### 3.5 Business uce Case

```python
#bar chart
resto_counts = df["location"].value_counts().head(10)
resto_counts.plot(kind="bar")
plt.title("Top 10 Lokasi dengan jumlah terbanyak")
plt.show()
```

untuk bisa melihat lokasi mana saja yang paling banyak ada restorannya.
dan ternyata wilayah yang paling banyak restorannya itu adalah "BTM/btm yang dimana persaingan bisnis kuliner yang sangat ketat dibandingkan wilayah lainnya.

```python
type_order_counts = df["online_order"].value_counts()

plt.figure()
type_order_counts.plot(
    kind='pie',
    autopct="%1.1f%%",
)
plt.ylabel("Online Order")
plt.title(" Proporsi Pesanan Online VS Offline")
plt.show()
```

Untuk melihat kesiapan digital restoran dan preferensi metode pemesanan (makan di tempat atau pesan antar).
Sebagian besar restoran (lebih dari 50%) sudah menyediakan layanan pemesanan online. Ini membuktikan bahwa restoran/kuliner sudah sangat bagus dengan teknologi digital.

```python
top_expensive_loc = df.groupby('location')['approx_cost(for two people)'].mean().sort_values(ascending=False).head(10)

plt.figure()
top_expensive_loc.plot(kind='bar')
plt.title('10 Lokasi dengan Rata-rata Biaya Makan Termahal')
plt.xlabel('Lokasi')
plt.ylabel('Rata-rata Harga')
plt.show()
```

untuk mengidentifikasi wilayah elit atau wilayah dengan biaya hidup (makan) paling tinggi.
dan ternyata Lokasi seperti sankey road atau race course road memiliki rata-rata harga tertinggi. Wilayah ini adalah target pasar potensial untuk restoran kelas atas.

```python
plt.figure()
sns.countplot(x='rest_type', data=df, order=df['rest_type'].value_counts().iloc[:10].index)
plt.xticks()
plt.title('10 Jenis Tipe Restoran yang Paling Banyak')
plt.show()
```

Memahami model bisnis restoran yang paling diminati oleh konsumen,
dan ternyata tipe quick bites adalah yang paling banyak diminnatti. Hal ini menunjukkan bahwa mereka lebih menyukai konsep makan yang tempat makannya santai untuk berkumpul.
